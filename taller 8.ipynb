2. Creamos la grafica realizada en clase de una funcion lineal, cuadrática y cubica

```python
x = np.linspace(0, 5, 100)


fig, ax = plt.subplots(figsize=(5, 5), layout='constrained')
ax.plot(x, x, label='lineal')
ax.plot(x, x**2, label='quadratica')
ax.plot(x, x**3, label='cubica')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
ax.set_title("GRAFICA DE FUNCIONES")
ax.legend()
```

1. Vamos a crear una figura que genere y visualice una distribución normal de datos aleatorios utilizando un histograma

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2), layout='constrained')

n, bins, patches = ax.hist(x, 50, density=True, facecolor='C0', alpha=0.75)

ax.set_xlabel('Longitud [cm]')
ax.set_ylabel('Probabilidad')
ax.set_title('Longitudes de Aardavark (en realidad no)')
ax.text(75, .025, r'$\mu=115,\ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True)
t = ax.set_xlabel('Mis Datos', fontsize=14, color='red')
```
